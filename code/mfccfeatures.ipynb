```python
import json
import os
import math
import librosa

sample_rate = 22050
track_duration = 30
track_samples = sample_rate * track_duration
dataset_path="genres_original"
json_path= "data.json"
num_mfcc=13
n_fft=2048
hop_length=512
num_segments=15
data =  {}
data['mappings']=[]
data['labels']=[]
data['mfcc']=[]
segment_samples = track_samples//num_segments
num_mfcc_vectors_per_segment =  math.ceil(segment_samples/hop_length)
```

```python
for val in enumerate(os.walk(dataset_path)):
    i = val[0]
    path = val[1][0]
    names = val[1][1]
    files = val[1][2]
    if path is not dataset_path:
        label = path.split('/')[-1]
        data['mappings'].append(label)
        print("\nProcessing: {}".format(label))
        for f in files:
            file_path = os.path.join(path, f)
            signal, sample_rate = librosa.load(file_path, sr=sample_rate)
            for d in range(num_segments):
                begin = segment_samples * d
                end = begin + segment_samples
                sig = signal[begin:end]
                mfcc = librosa.feature.mfcc(sig, sample_rate, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
                mfcc = mfcc.T
                if len(mfcc) == num_mfcc_vectors_per_segment:
                    data["mfcc"].append(mfcc.tolist())
                    data["labels"].append(i-1)
                    print("{}, segment:{}".format(file_path, d+1))
                    
                    
```

```python
with open(json_path, "w") as fp:
    json.dump(data, fp, indent=4)
```
